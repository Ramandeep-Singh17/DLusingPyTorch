# ðŸ§  Tensors in PyTorch

---

## ðŸ”¹ What is a Tensor?

Tensor is a **data structure used to store numerical data in Deep Learning**.

In simple words:
> Tensor = Multi-dimensional array

It can store:
- Numbers
- Vectors
- Images
- Videos
- Model weights
- Predictions

PyTorch ka main building block = Tensor

---

## ðŸ”¹ Why Do We Use Tensors?

We use tensors because:

- Neural networks work only on numbers
- GPU computation needs structured numeric data
- Matrix operations (multiplication, addition) need multi-dimensional arrays
- Backpropagation needs tensors to compute gradients

Without tensors:
- No forward pass
- No loss calculation
- No training

---

## ðŸ”¹ Where Are Tensors Used?

Tensors are used in:

- Images (Computer Vision)
- Text (NLP)
- Audio processing
- Videos
- Model parameters (weights)
- Loss values
- Gradients

Every input and output in Deep Learning = Tensor

---

## ðŸ”¹ When Do We Use Tensors?

We use tensors:

- While giving input to model
- During forward pass
- While calculating loss
- During backpropagation
- During prediction

---

## ðŸ”¹ How Does Tensor Represent Data?

Tensor stores data based on **dimensions (axes)**.

---

# ðŸ“Š Types of Tensors (Based on Dimensions)

---

## 0ï¸âƒ£ 0D Tensor (Scalar)

- Single number
- No direction

Example:
Loss value

If:
Prediction = 8  
Target = 10  
Loss = 2  

That 2 is a scalar (0D tensor)

âœ” Loss is always scalar

---

## 1ï¸âƒ£ 1D Tensor (Vector)

- One direction
- Like a list or array

Example:
[2, 5, 7, 9]

### ðŸ“Œ Example: Word Embedding

Embedding means:
> Converting words into numbers (vectors)

Example:
"king"  â†’ [0.25, 0.71, 0.90]  
"queen" â†’ [0.27, 0.69, 0.88]  

Similar words â†’ similar vectors

Used in:
- NLP
- Chatbots
- LLMs

Embedding = 1D tensor

---

## 2ï¸âƒ£ 2D Tensor (Matrix)

- Two directions (rows, columns)

Example:
[[1, 2, 3],
 [4, 5, 6]]

### ðŸ“Œ Example: Grayscale Image

Grayscale image has:
- Height
- Width

Shape:
[Height, Width]

Example:
[28, 28]

---

## 3ï¸âƒ£ 3D Tensor

- Three directions

### ðŸ“Œ Example: RGB Image

RGB image has:
- Channels (3)
- Height
- Width

Shape:
[Channels, Height, Width]

Example:
[3, 224, 224]

---

## 4ï¸âƒ£ 4D Tensor

- Four directions

### ðŸ“Œ Example: Batch of RGB Images

During training, model takes multiple images together.

Shape:
[Batch_size, Channels, Height, Width]

Example:
[32, 3, 224, 224]

Meaning:
32 images in one batch

---

## 5ï¸âƒ£ 5D Tensor

- Five directions

### ðŸ“Œ Example: Video

Video = Sequence of images

Shape:
[Batch_size, Time, Channels, Height, Width]

Example:
[8, 10, 3, 224, 224]

Meaning:
8 videos  
Each video has 10 frames

| **16-bit Floating Point** | `torch.float16`   | Half-precision floating point. Commonly used in mixed-precision training to reduce memory and computational overhead on modern GPUs.                                            |
| **BFloat16**              | `torch.bfloat16`  | Brain floating-point format with reduced precision compared to `float16`. Used in mixed-precision training, especially on TPUs.                                                |
| **8-bit Floating Point**  | `torch.float8`    | Ultra-low-precision floating point. Used for experimental applications and extreme memory-constrained environments (less common).                                               |
| **8-bit Integer**         | `torch.int8`      | 8-bit signed integer. Used for quantized models to save memory and computation in inference.                                                                                   |
| **16-bit Integer**        | `torch.int16`     | 16-bit signed integer. Useful for special numerical tasks requiring intermediate precision.                                                                                    |
| **32-bit Integer**        | `torch.int32`     | Standard signed integer type. Commonly used for indexing and general-purpose numerical tasks.                                                                                  |
| **64-bit Integer**        | `torch.int64`     | Long integer type. Often used for large indexing arrays or for tasks involving large numbers.                                                                                  |
| **8-bit Unsigned Integer**| `torch.uint8`     | 8-bit unsigned integer. Commonly used for image data (e.g., pixel values between 0 and 255).                                                                                    |
| **Boolean**               | `torch.bool`      | Boolean type, stores `True` or `False` values. Often used for masks in logical operations.                                                                                      |
| **Complex 64**            | `torch.complex64` | Complex number type with 32-bit real and 32-bit imaginary parts. Used for scientific and signal processing tasks.                                                               |
| **Complex 128**           | `torch.complex128`| Complex number type with 64-bit real and 64-bit imaginary parts. Offers higher precision but uses more memory.                                                                 |
| **Quantized Integer**     | `torch.qint8`     | Quantized signed 8-bit integer. Used in quantized models for efficient inference.                                                                                              |

## Mathematical operations

### 1. Scalar operation

````python
x = torch.rand(2,2)
x
````

````python
# addition
x + 2
# substraction
x - 2
# multiplication
x * 3
# division
x / 3
# int division
(x * 100)//3
# mod
((x * 100)//3)%2
# power
x**2
````

### 2. Element wise operation

````python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
````

````python
# add
a + b
# sub
a - b
# multiply
a * b
# division
a / b
# power
a ** b
# mod
a % b
````

````python
# mean
torch.mean(e)
# mean along col
torch.mean(e, dim=0)
````

````python
# median
torch.median(e)
````

### 4. Matrix operations

````python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
````

````python
# matrix multiplcation
torch.matmul(f, g)
````

````python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
````

````python
# transpose
torch.transpose(f, 0, 1)
````

````python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
````

````python
# determinant for sq matrx only
torch.det(h)
````

````python
# inverse # V for sq matrx onlytorch.inverse(h)
````

### 5. Comparison operations

````python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
````

````python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to

# less than equal to
````

## ðŸ”¹ Sigmoid  
Sigmoid is an activation function that converts any input value into a range between 0 and 1.

## ðŸ”¹ ReLU  
ReLU (Rectified Linear Unit) is an activation function that outputs 0 for negative values and keeps positive values unchanged.

## ðŸ”¹ Softmax  
Softmax is an activation function that converts multiple values into probability distribution (sum = 1).


````python
# sigmoid
torch.sigmoid(k)
````

````python
# softmax
torch.softmax(k, dim=0)
````

````python
# relu
torch.relu(k)
````
